```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for class_id in range(1,11):
    print(f"Class {class_id} images:")
```

```python
dt = pd.read_csv('Tools_script/ball_counting_dataset_all.csv')
```

```python
column_data = dt['ball_count']
```

```python
label_counts = column_data.value_counts().sort_index()

# 设置图表样式
sns.set_palette("husl")

# 创建图表
plt.figure(figsize=(12, 6))
bars = plt.bar(label_counts.index, label_counts.values, width=0.6)

# 添加标题和标签
plt.title('Distribution of Labels', fontsize=20, fontweight='bold')
plt.xlabel('Labels', fontsize=14)
plt.ylabel('Count', fontsize=14)

# 在每个柱子上方添加数值标签
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom', fontsize=11)

# 调整x轴刻度
plt.xticks(range(1, 11))

# 添加网格线
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 调整布局并显示图表
plt.tight_layout()
plt.show()
```
